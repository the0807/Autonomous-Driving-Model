```python
import os
import json
import shutil
```

```python
# 변수 선언

# 데이터셋의 루트 디렉토리 설정
root_dir = "dataset/2DSS"
preprocessed_root_dir = "dataset/Preprocessed_2DSS"

# 하위 디렉토리들
sub_dirs = ["training", "validation", "test"]

data_num = 0
data_num_count = 0

class_mapping = {
    "road": 0, 
    "sidewalk": 1, 
    "road roughness": 2, 
    "road boundaries": 3, 
    "crosswalks": 4, 
    "lane": 5, 
    "road color guide": 6, 
    "road marking": 7, 
    "parking": 8, 
    "traffic sign": 9, 
    "traffic light": 10, 
    "pole/structural object": 11, 
    "building": 12, 
    "tunnel": 13, 
    "bridge": 14, 
    "pedestrian": 15, 
    "vehicle": 16, 
    "bicycle": 17, 
    "motorcycle": 18, 
    "personal mobility": 19, 
    "dynamic": 20, 
    "vegetation": 21, 
    "sky": 22, 
    "static": 23
}

coco_categories = [
    {'id': 0, 'name': 'road', 'supercategory': 'flat'},
    {'id': 1, 'name': 'sidewalk', 'supercategory': 'flat'},
    {'id': 2, 'name': 'road roughness', 'supercategory': 'flat'},
    {'id': 3, 'name': 'road boundaries', 'supercategory': 'flat'},
    {'id': 4, 'name': 'crosswalks', 'supercategory': 'flat'},
    {'id': 5, 'name': 'lane', 'supercategory': 'flat'},
    {'id': 6, 'name': 'road color guide', 'supercategory': 'flat'},
    {'id': 7, 'name': 'road marking', 'supercategory': 'flat'},
    {'id': 8, 'name': 'parking', 'supercategory': 'flat'},
    {'id': 9, 'name': 'traffic sign', 'supercategory': 'static object'},
    {'id': 10, 'name': 'traffic light', 'supercategory': 'static object'},
    {'id': 11, 'name': 'pole/structural object', 'supercategory': 'static object'},
    {'id': 12, 'name': 'building', 'supercategory': 'construction'},
    {'id': 13, 'name': 'tunnel', 'supercategory': 'construction'},
    {'id': 14, 'name': 'bridge', 'supercategory': 'construction'},
    {'id': 15, 'name': 'pedestrian', 'supercategory': 'moving object'},
    {'id': 16, 'name': 'vehicle', 'supercategory': 'moving object'},
    {'id': 17, 'name': 'bicycle', 'supercategory': 'moving object'},
    {'id': 18, 'name': 'motorcycle', 'supercategory': 'moving object'},
    {'id': 19, 'name': 'personal mobility', 'supercategory': 'moving object'},
    {'id': 20, 'name': 'dynamic', 'supercategory': 'moving object'},
    {'id': 21, 'name': 'vegetation', 'supercategory': 'nature'},
    {'id': 22, 'name': 'sky', 'supercategory': 'nature'},
    {'id': 23, 'name': 'static', 'supercategory': 'static'},
]
```

```python
# dataset 디렉토리 복사

shutil.copytree(root_dir, preprocessed_root_dir)
```

```python
# 이미지와 json 결측 파일 확인 후 삭제

for sub_dir in sub_dirs:
    label_dir = os.path.join(preprocessed_root_dir, sub_dir, 'labels')
    image_dir = os.path.join(preprocessed_root_dir, sub_dir, 'images')
    
    label_filenames = os.listdir(label_dir)
    image_filenames = os.listdir(image_dir)

    missing_images = [filename for filename in label_filenames if os.path.splitext(filename)[0] + '.jpg' not in image_filenames]
    missing_json = [filename for filename in image_filenames if os.path.splitext(filename)[0] + '.json' not in label_filenames]
    
    print(f"디렉토리: {sub_dir}")
    print("삭제할 이미지:", missing_images)
    print("삭제할 json:", missing_json)
    
    file_num = 0
    for filename in missing_images:
        file_path = os.path.join(label_dir, filename)
        os.remove(file_path)
        file_num += 1
    
    for filename in missing_json:
        file_path = os.path.join(image_dir, filename)
        os.remove(file_path)
        file_num += 1

    print(f"총 {file_num}개의 파일을 삭제했습니다.")
    print('')
```

```python
# 디렉토리 안의 파일 개수 확인

for sub_dir in sub_dirs:
    images_dir = os.path.join(preprocessed_root_dir, sub_dir, "images")
    labels_dir = os.path.join(preprocessed_root_dir, sub_dir, "labels")
    
    images_count = len(os.listdir(images_dir))
    labels_count = len(os.listdir(labels_dir))
    
    print(f"디렉토리: {sub_dir}")
    print(f"images 개수: {images_count}")
    print(f"labels 개수: {labels_count}")
    print('')
```

```python
# json 데이터셋 coco 형식으로 변환하는 함수 정의

def convert_to_coco(json_data, image_id, annotation_id_start, category_mapping):
    images = []
    annotations = []
    current_annotation_id = annotation_id_start
    
    image_info = {
        "id": image_id,
        "file_name": json_data["image_name"],
        "width": json_data["image_size"][0],
        "height": json_data["image_size"][1]
    }
    images.append(image_info)

    for annotation in json_data["Annotation"]:
        class_name = annotation["class_name"]
        category_id = category_mapping[class_name]

        for polygon in annotation["data"]:
            segmentation = [polygon]

            x_coordinates = polygon[::2]
            y_coordinates = polygon[1::2]
            xmin = min(x_coordinates)
            xmax = max(x_coordinates)
            ymin = min(y_coordinates)
            ymax = max(y_coordinates)
            width = xmax - xmin
            height = ymax - ymin
            bbox = [xmin, ymin, width, height]

            annotation_info = {
                "id": current_annotation_id,
                "image_id": image_id,
                "category_id": category_id,
                "segmentation": segmentation,
                "area": width * height,
                "bbox": bbox,
                "iscrowd": 0
            }
            annotations.append(annotation_info)
            current_annotation_id += 1

    image_id += 1

    return {
        "images": images,
        "annotations": annotations
    }
```

```python
# json 데이터셋 coco 형식으로 변환

for sub_dir in sub_dirs:
    label_dir = os.path.join(preprocessed_root_dir, sub_dir, 'labels')
    coco_dir = os.path.join(preprocessed_root_dir, sub_dir, 'coco')
    
    label_filenames = [os.path.join(label_dir, filename) for filename in os.listdir(label_dir)]
    
    output_coco_file = coco_dir + f'/{sub_dir}_coco.json'
    os.makedirs(coco_dir, exist_ok=True)

    all_images = []
    all_annotations = []
    
    image_id = 0
    annotation_id_start = 0
    
    for file_path in label_filenames:
        with open(file_path, 'r', encoding='utf-8') as f:
            json_data = json.load(f)
            coco_data = convert_to_coco(json_data, image_id, annotation_id_start, class_mapping)
            all_images.extend(coco_data["images"])
            all_annotations.extend(coco_data["annotations"])
            image_id += len(coco_data["images"])
            annotation_id_start += len(coco_data["annotations"])
    
    coco_format = {
        "images": all_images,
        "annotations": all_annotations,
        "categories": coco_categories
    }
    
    with open(output_coco_file, 'w', encoding='utf-8') as f:
        json.dump(coco_format, f, ensure_ascii=False, indent=4)
```

```python
# coco 형식으로 디렉토리 이동

for sub_dir in sub_dirs:
    label_dir = os.path.join(preprocessed_root_dir, sub_dir, 'labels')
    coco_label_dir = os.path.join(preprocessed_root_dir, sub_dir, 'coco')
    image_dir = os.path.join(preprocessed_root_dir, sub_dir, 'images')
    new_label_dir = os.path.join(preprocessed_root_dir, 'annotations', sub_dir)
    new_image_dir = os.path.join(preprocessed_root_dir, 'images', sub_dir)
    
    os.makedirs(new_label_dir, exist_ok=True)
    os.makedirs(new_image_dir, exist_ok=True)

    for filename in os.listdir(image_dir):
        source_image_path = os.path.join(image_dir, filename)
        target_image_path = os.path.join(new_image_dir, filename)

        shutil.copyfile(source_image_path, target_image_path)
        
    for filename in os.listdir(coco_label_dir):
        source_label_path = os.path.join(coco_label_dir, filename)
        target_label_path = os.path.join(new_label_dir, filename)

        shutil.copyfile(source_label_path, target_label_path)
        
    shutil.rmtree(image_dir)
    shutil.rmtree(label_dir)
    shutil.rmtree(coco_label_dir)
    shutil.rmtree(os.path.join(preprocessed_root_dir, sub_dir))
```
